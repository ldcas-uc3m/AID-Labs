```python
from PIL import Image

# Carga la imagen `Arborio (1).jpg` con la funci√≥n Image del paquete PIL
input_dir = 'data/Rice_Image_Dataset'

image = Image.open(f'{input_dir}/Arborio/Arborio (1).jpg')
image
```

```python
from torchvision import transforms

# Crea un transformador de Resize de 128x128
resize_transform = transforms.Resize((128, 128))
# Utiliza el transformador sobre la imagen `image`
resized_image = resize_transform(image)
resized_image
```

```python
# Crea un transformador de CenterCrop
centercrop_transform = transforms.CenterCrop(500)
# Utiliza el transformador sobre la imagen `image`
centercropped_image = centercrop_transform(image)
centercropped_image
```

```python
# Crea un transformador de RandomHorizontalFlip
randomhorizontalflip_transform = transforms.RandomHorizontalFlip(p=0.9)
# Utiliza el transformador sobre la imagen `image`
flipped_image = randomhorizontalflip_transform(image)
flipped_image
```

```python
# Crea un transformador de RandomRotation
randomrotation_transform = transforms.RandomRotation(degrees=15)
# Utiliza el transformador sobre la imagen `image`
rotated_image = randomrotation_transform(image)
rotated_image
```

```python
import matplotlib.pyplot as plt

def show_images(images, titles):
    plt.figure(figsize=(15, 10))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(img)
        plt.title(title)
        plt.axis('off')
    plt.show()

show_images(
    [resized_image, centercropped_image, flipped_image, rotated_image],
    ['Resize', 'CenterCrop', 'RandomHorizontalFlip', 'RandomRotation']
)
```

# Linear Regression

```python
from sklearn.datasets import load_diabetes

diabetes_dataset = load_diabetes()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    diabetes_dataset.data,
    diabetes_dataset.target,
    test_size=0.2,
    random_state=1234
)
```

```python
# Crear un pipeline con un StandardScaler y el modelo LinearRegression, y entrenarlo
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
# Obtener las predicciones y el MSE (Mean Squared Error)
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse}")
```

```python
import matplotlib.pyplot as plt

# Muestra una gráfica de puntos con matplotlib siguiendo las siguientes instrucciones:
# - El eje X debe representar los valores reales
# - El eje Y debe representar las predicciones
# - Debe aparecer una línea discontinua indicando donde deberían estar los puntos si el modelo fuera ideal
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, label='Predicciones')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Ideal')
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Predicciones vs. Valores Reales')
plt.legend()
plt.show()
```

```python
# Obten los residuos, es decir, la diferencia entre el valor real y el valor de la predicción
residuos = y_test - y_pred

# Muestra una gráfica de puntos siguiendo las siguientes instrucciones:
# - El eje X debe representar las predicciones
# - El eje Y debe representar los residuos
# - Dibuja una línea discontinua horizontal con valor en el eje Y de 0. Esta línea representa el residuo 0
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuos, alpha=0.5)
plt.axhline(y=0, color='k', linestyle='--')
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos')
plt.show()
```
